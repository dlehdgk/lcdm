# loading modules

```python
import numpy as np
import matplotlib.pyplot as plt
import getdist
from getdist import plots
```

# loading the chains

```python
# need to explicitly indicate the repository containing the chain file
chain_root = "./lcdm"

samples = getdist.loadMCSamples(chain_root, settings={"ignore_rows": 0.3})
```

# testing convergence

```python
print("R-1=", samples.getGelmanRubin())
```

# constaints on parameters

```python
# 68% CL = 1 sigma
# As
print(samples.getInlineLatex("As", 1))

# ns
print(samples.getInlineLatex("ns", 1))

# ombh2
print(samples.getInlineLatex("ombh2", 1))

# omch2
print(samples.getInlineLatex("omch2", 1))

# tau
print(samples.getInlineLatex("tau", 1))

# H0
print(samples.getInlineLatex("H0", 1))
```

```python
# triangle plots showing 1D posterior distribution of parameters and correlation between
# parameters using marginalised contour plots
g = getdist.plots.get_subplot_plotter(width_inch=8)
params = ["ombh2", "omch2", "tau", "H0", "ns", "As"]
g.triangle_plot(samples, params, filled=True)
```
